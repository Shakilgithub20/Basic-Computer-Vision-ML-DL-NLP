```
def read_files(file_loc):
  dataset = []
  
  with open(file_loc, 'r', encoding='cp1258') as test_file:
    for line in test_file:
      dataset.append(line)
    
  return dataset
```

```
# function to Tokenization
def tokenization(data):
  tokenization_data = []

  for line in data:
    tokenization_data.append(word_tokenize(line))

    return tokenization_data
```

```
# function to Text lower case
def lowercase(data):
    lowercase_data = []

    for line in data:
      new_word = line.lower()
      lowercase_data.append(new_word)
      return lowercase_data
```

```
def separate_labels(dataset):
  documents = []
  labels = []

  for line in dataset:
    splitted_line = line.strip().split('\t', 2)
    labels.append(splitted_line[1])
    documents.append(splitted_line[2])

  return documents, labels
```

```
def remove_hashtag(data):
  '''This function removes HashTag # from the entire dataset'''
  hashtag_removed_data = []

  # map hashtag to space
  translator = str.maketrans('#', ' '*len('#'), '')

  for line in data:
    hashtag_removed_data.append(line.translate(translator))

  return hashtag_removed_data
```

```
def remove_urls(data):
  url_removed_data = []

  # Your Code Here...
  for line in data:
    url_removed_data.append(re.sub('http[s]?://\S+', '', line))

  return url_removed_data
```

```
def remove_urls(data):
  url_removed_data = []

  # Your Code Here...
  for line in data:
    url_removed_data.append(re.sub('http[s]?://\S+', '', line))

  return url_removed_data
```

```
# Remove numbers 
def remove_numbers(data):

  numbers_removed_data = []

  for line in data:
    numbers_removed_data.append(re.sub('[0-9]+', '', line))

    return  numbers_removed_data 
```

```
# Expand contracted data
def expanded_contracted(data):
    """Replace contractions in string of text"""
    expanded_words_data = []
    
    for line in data:
      expanded_words_data.append(contractions.fix(line))
      
      return expanded_words_data
```

```
# function to remove special characters
def remove_special_characters(data):
    # define the pattern to keep
    remove_special_characters_data = []
    
    pat = r'[^a-zA-z0-9.,!?/:;\"\'\s]' 

    for line in data:
      remove_special_characters_data.append(re.sub(pat, '', line))
      
      return remove_special_characters_data
```

```
# function to remove extra whitespace tabs
def remove_extra_whitespace_tabs(data):
  remove_extra_whitespace_tabs_data = []
  pattern = r'^\s*|\s\s*'
  for line in data:
    remove_extra_whitespace_tabs_data.append(re.sub(pattern, ' ', line).strip())
  return remove_extra_whitespace_tabs_data
```

```
# function to remove punctuation 
def remove_punctuation(data):
  remove_punctuation_data = []
  
  translator = str.maketrans('', '', string.punctuation)
  
  for line in data:
      remove_punctuation_data.append(line.translate(translator))
      
      return remove_punctuation_data
```

```
def pre_processing(documents):
  '''This function calls all the pre-processing submodules'''

  documents = remove_urls(documents)
  documents = remove_hashtag(documents)
  documents = remove_whitespace(documents)
  documents = remove_numbers(documents)
  documents = expanded_contracted(documents)
  documents = remove_special_characters(documents)
  documents = lowercase(documents)
  documents = remove_extra_whitespace_tabs(documents)
  documents = remove_punctuation(documents)
  documents = tokenization(documents)

  return documents
```

```
def flatten(x):
    result = []
    for el in x:
        if hasattr(el, "__iter__") and not isinstance(el, str):
            result.extend(flatten(el))
        else:
            result.append(el)
    return result
```
